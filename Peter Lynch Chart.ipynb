```python
def get_lynch_chart(exchange, stock):
    xls = pd.ExcelFile("{}/API/{}/annual.xlsx".format(exchange, stock))
    df = xls.parse('income_statement')
    df = df.rename(columns = {'TTM':'2019-06'})
    df.columns = pd.to_datetime(df.columns)
    df_EPS = df.loc['net_income'] / df.loc['shares_basic']
    start_date = '1999-09-01'
    end_date = '2019-04-01'
    panel_data = pdr.DataReader(stock, 'yahoo', start_date, end_date)
    trace1 = go.Scatter(
        x = df.columns,
        y =  35 * df_EPS,
        name = "Earnings",
    )

    trace2 = go.Scatter(
        x = panel_data.index,
        y = panel_data['Close'],
        name = "Stock Price",
    )



    data = [trace1, trace2]
    layout = dict(
        title = 'test'
    )

    fig = dict(data = data, layout = layout)
    iplot(fig, filename = 'test2')
```

```python
files = os.listdir(cwd+ "/NASDAQ/API")

def check_stock_15_growth(stock):
    print(stock)
    xls = pd.ExcelFile("NASDAQ/API/{}/annual.xlsx".format(stock))
    df = xls.parse('income_statement')
    if len(df.columns) >= 6:
        df = df.rename(columns = {'TTM':'2019-06'})
        df.columns = pd.to_datetime(df.columns)
        change = df.loc['net_income'][-6:].pct_change()*100
        hypothesis = True
        for val in change[1:]:
            if val >= 15:
                hypothesis = hypothesis * True
            else:
                hypothesis = hypothesis * False
        return hypothesis
    else:
        print("Stock {} does not have enough data".format(stock))
        return False
    
stalemate_stocks = []
for ticker in files:
    print("checking {}".format(ticker))
    pd = check_stock_15_growth(ticker)
    if pd:
        stalemate_stocks.append(ticker)
    else:
        pass
```
